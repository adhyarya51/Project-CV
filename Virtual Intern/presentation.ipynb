## data load

```python
# loading dataset 

data_virtue = pd.read_csv('Data (1).csv', sep=';') 

#  duplikate original data to make things easier. 

df = data_virtue.copy()
```

```python
df
```

```python
# Extract the repayment columns (all columns except the first two)
repayment_columns = df.columns[2:]

# Calculate repayment percentages
for col in repayment_columns:
    df[f'{col}_percentage'] = df[col] / df['Origination Amount']
    
df.head()
```

```python
# Calculate average historical repayment percentages
historical_repayment_percentages = df[[f'{col}_percentage' for col in repayment_columns]].mean()

# Convert to a dictionary for easy access
expected_repayments = historical_repayment_percentages.to_dict()

# Display the expected repayment percentages
expected_repayments
```
