##### Statistik Inferensial

didalam statistik inferensial saya akan menggunakan confidence interval.

````python
dataCleaned = pd.DataFrame(dataCleaned)

# Sub-Saharan Africa          
# Europe                      
# Asia-Pacific                
# Americas                    
# Middle East/North Africa 

SubSaharanOverall = dataCleaned[dataCleaned['Region']=='Sub-Saharan Africa']['Overall Score']
EuropeOverall = dataCleaned[dataCleaned['Region']=='Europe']['Overall Score']
AsiaPacificOverall = dataCleaned[dataCleaned['Region']=='Asia-Pacific']['Overall Score']
AmericasOverall = dataCleaned[dataCleaned['Region']=='Americas']['Overall Score']
MiddleEastNorthAfricaOverall = dataCleaned[dataCleaned['Region']=='Middle East/North Africa']['Overall Score']

#dalam part ini saya mencoba melakukan perhitungan dalam pengelompokan

mean_sub = np.mean(SubSaharanOverall)
std_sub = np.std(SubSaharanOverall)

mean_euro = np.mean(EuropeOverall)
std_euro = np.std(EuropeOverall)

mean_asia = np.mean(AsiaPacificOverall)
std_asia = np.std(AsiaPacificOverall)

mean_america = np.mean(AmericasOverall)
std_america = np.std(AmericasOverall)

mean_middle = np.mean(MiddleEastNorthAfricaOverall)
std_middle = np.std(MiddleEastNorthAfricaOverall)


#perhitungan saya lakukan masing - masing kelompok dan keseluruhan dengan tingkat kepercayaan 95%
conf_int_sub = stats.norm.interval(0.95,loc=mean_sub,scale=std_sub/np.sqrt(len(SubSaharanOverall)))
conf_int_euro = stats.norm.interval(0.95,loc=mean_euro,scale=std_euro/np.sqrt(len(EuropeOverall)))
conf_int_asia = stats.norm.interval(0.95,loc=mean_asia,scale=std_asia/np.sqrt(len(AsiaPacificOverall)))
conf_int_america = stats.norm.interval(0.95,loc=mean_america,scale=std_america/np.sqrt(len(AmericasOverall)))
conf_int_middle = stats.norm.interval(0.95,loc=mean_middle,scale=std_middle/np.sqrt(len(MiddleEastNorthAfricaOverall)))

print('Confidence Interval dalam pengukuran region sub saharan africa : ', conf_int_sub)
print('Confidence Interval dalam pengukuran region eropa :', conf_int_euro)
print('Confident Interval dalam pengukuran region asia pacific :', conf_int_asia)
print('Confident Interval dalam pengukuran region america :', conf_int_america)
print('Confident Interval dalam pengukuran region middle east and north africa :', conf_int_middle)
````

Dari hasil confidence interval untuk setiap kolom perhitungan, dapat disimpulkan menjadi beberapa yaitu : 

1. Sub Saharan Africa : Confidence interval untuk rata rata region antara 50,399 dan 54,511 dengan tingkat kepercayaan 95%.

2.  Europe : Confidence interval untuk rata rata region antara 64,872 dan 69,504 dengan tingkat kepercayaan 95%.

3.  Asia Pacific : Confidence interval untuk rata rata region antara 53,34 dan 61,52 dengan tingkat kepercayaan 95%.

4.  America : Confidence interval untuk rata rata region antara 54,285 dan 61,708 dengan tingkat kepercayaan 95%.

5.  middle east and north africa : Confidence interval untuk rata rata region antara 52,500 dan 62,314 dengan tingkat kepercayaan 95%.

Dari hasil tersebut dilihat masing - masing region memiliki nilai yang mendekati satu sama lain, dengan region tertinggi pada wilayah europe. kita memiliki keyakinan bahwa setiap region memiliki kisaran confidence interval yang diberikan.

##### Hypothesis Testing

pada bagian ini saya menggunakan metode ```two simple independent test``` dengan melakukan perbandingan antara negara ASEAN dan negara non ASEAN.
Test tersebut bertujuan untuk mengetes dua kondisi. saya menggunakan modul ```scipy.stats```, dan akan melakukan pengukuran pada nilai p-value dengan standart perhitungan 0,05.

- H0 : nilai overall score negara ASEAN == nilai overall score negara non ASEAN
- H1 : nilai overall score negara ASEAN != nilai overall score negara non ASEAN 

````python
#melakukan pemindahan data csv untuk mempermudah pemanggilan
dataNonASEAN = pd.read_csv('Freedom_index(non ASEAN)')
dataASEAN = pd.read_csv('ASEAN')
````

````python
print('Rata - rata overall score negara nonASEAN =',dataNonASEAN['Overall Score'].mean())
print('Rata - rata overall score negara ASEAN =',dataASEAN['Overall Score'].mean())
````

````python
#perhitungan T statistic dan P value
t_stat, p_val = stats.ttest_ind(dataASEAN['Overall Score'],dataNonASEAN['Overall Score'])
print('T-statistic:',t_stat)
print('P-value',p_val)
````

Dari hasil yang ditemukan bahwa score didapatkan memiliki keterhubungan satu sama lain, dan memiliki hubungan secara positif dengan dibuktikan hasil data p-value > 0.05.
